```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as spm
import math
%matplotlib inline
```

```python
#課題1

n=50
p=0.7
n_rand=100000
rand_binominal=np.random.binomial(n,p,n_rand)

def binominal(n,p,x):
    return spm.comb(n,x)*(p**x)*((1-p)**(n-x))

#ヒストグラム
plt.hist(rand_binominal,bins=range(rand_binominal.max()+1),alpha=0.5,color='blue',normed=True)

#pdfのプロット
x=np.array(range(rand_binominal.max()+1))
y_binominal=binominal(n,p,x)
plt.plot(x+0.5,y_binominal,color='blue')
plt.legend(['binominal'])
```

```python
#課題2

lam=p*n
rand_poisson=np.random.poisson(lam,n_rand)
def poisson(lam,x):
    return lam**x/map(math.factorial,x)*np.exp(-lam)

#ヒストグラム
a=plt.hist(rand_poisson,bins=range(rand_poisson.max()+1),alpha=0.5,color='red',normed=True)

#pdfのプロット
x=np.array(range(rand_poisson.max()+1))
y_poisson=poisson(n*p,x)
plt.plot(x+0.5,y_poisson,color='red')
plt.legend(['poisson'])
```

```python
#課題3

n=5000
p=0.007
n_rand=100000
rand_binominal=np.random.binomial(n,p,n_rand)

def binominal(n,p,x):
    return spm.comb(n,x)*(p**x)*((1-p)**(n-x))

#ヒストグラム
plt.hist(rand_binominal,bins=range(rand_binominal.max()+1),alpha=0.5,color='blue',normed=True)

#pdfのプロット
x=np.array(range(rand_binominal.max()+1))
y_binominal=binominal(n,p,x)
plt.plot(x+0.5,y_binominal,color='blue')
plt.legend(['binominal'])
```

```python
#課題4

lam=p*n
rand_poisson=np.random.poisson(lam,n_rand)

#ヒストグラム
a=plt.hist(rand_poisson,bins=range(rand_poisson.max()+1),alpha=0.5,color='red',normed=True)

#pdfのプロット
x=np.array(range(rand_poisson.max()+1))
y_poisson=poisson(n*p,x)
plt.plot(x+0.5,y_poisson,color='red')
plt.legend(['poisson'])
```

```python
#課題5

plt.hist(rand_binominal,bins=range(rand_binominal.max()+1),alpha=0.5,color='blue',normed=True)
plt.hist(rand_poisson,bins=range(rand_poisson.max()+1),alpha=0.5,color='red',normed=True)

x=np.array(range(rand_poisson.max()+1))
y_binominal=binominal(n,p,x)
plt.plot(x+0.5,y_binominal,color='blue')
y_poisson=poisson(n*p,x)
plt.plot(x+0.5,y_poisson, color='red')
plt.legend(['binominal','poisson'])
```

```python
#課題6

# ポアソン過程

# T=n*dtの間をn個に分割
# スパイク数の期待値はn*dt*r
n=100000
r=0.1
dt=0.01

spike_poisson=np.random.binomial(1,r*dt,n)
print spike_poisson
plt.plot(spike_poisson)
```

```python
#課題7

#複数のビンに区切る
spike_timing_poisson=np.where(spike_poisson==1)
firing_rate_poisson=plt.hist(spike_timing_poisson,10)

isi_poisson=np.diff(spike_timing_poisson)
cv_poisson=isi_poisson.std()/isi_poisson.mean()
fano_poisson=firing_rate_poisson[0].var()/firing_rate_poisson[0].mean()

print('cv=   '+str(cv_poisson))
print('Fano= '+str(fano_poisson))
```

```python
#課題7
spike_regular=np.zeros(n)
interval=int(n/(r*dt*n))
ind=range(0,n,interval)
spike_regular[ind]=1
plt.plot(spike_regular)
```

```python
#課題8

spike_timing_regular=np.where(spike_regular==1)
firing_rate_regular=plt.hist(spike_timing_regular,100)

isi_regular=np.diff(spike_timing_regular)
cv_regular=isi_regular.std()/isi_regular.mean()

fano_regular=firing_rate_regular[0].var()/firing_rate_regular[0].mean()

print('cv=   '+str(cv_regular))
print('Fano= '+str(fano_regular))
```
