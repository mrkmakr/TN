```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats.distributions as dis
%matplotlib inline
```

```python
#課題1

#正規分布を定義
def func(x,m,sig):
    return np.exp(-(x-m)**2/2/sig**2)/np.sqrt(2*np.pi)/sig

#自分で定義した分布をプロット
m=0
sig=1
range_x=10

x=np.linspace(-range_x,range_x,1000)
plt.plot(x,func(x,m,sig))
```

```python
#課題2

#複数のパラメータでのプロット
for k in range(1,5):
    m=0
    sig=0.9*k
    
    y=func(x,m,sig)
    plt.plot(x,y,label="sig="+str(sig))
plt.legend()
```

```python
#課題3

#乱数生成
sig=1
m=0
rand=np.random.normal(m,sig,10000)
#ヒストグラムのプロット
his=plt.hist(rand,100,normed=True)
plt.xlim([-10,10])
```

```python
#課題4

#組み込み関数による正規分布
plt.plot(x,dis.norm.pdf(x,loc=m,scale=sig))
```

```python
#課題5

#同時にプロット、重なる
plt.plot(x,func(x,m,sig))
plt.plot(x,dis.norm.pdf(x,loc=m,scale=sig))
```
