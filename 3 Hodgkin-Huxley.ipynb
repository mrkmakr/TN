```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def calculate_m(v):
    ma=-0.1*(50+v)/(np.exp(-(v+50)/10)-1)
    mb=4*np.exp(-(v+75)/18)
    mt=1/(ma+mb)
    minf=ma*mt
    return minf,mt
```

```python
def calculate_h(v):
    ha=0.07*np.exp(-(v+75)/20)
    hb=1/(np.exp(-(v+45)/10)+1)
    ht=1/(ha+hb)
    hinf=ha*ht
    return hinf,ht
```

```python
def calculate_n(v):
    na=-0.01*(v+65)/(np.exp(-(v+65)/10)-1)
    nb=0.125*np.exp(-(v+75)/80)
    nt=1/(na+nb)
    ninf=na*nt
    return ninf,nt
```

```python
class HHmodel():
    def __init__(self):
        self.cm=1#uF/cm^2
        
        self.gna=120#mS/cm^2
        self.gk=36
        self.gl=0.3

        self.ena=40#mV
        self.ek=-87
        self.el=-64.387

        self.dt=0.01#ms
        self.T=100
        self.step_number=int(self.T/self.dt)    
        
        self.amp_stim=20#uA/cm^2
        
        self.v=-75.0
        self.m,_=calculate_m(-75.0)
        self.h,_=calculate_h(-75.0)
        self.n,_=calculate_n(-75.0)
        
        self.stim=np.zeros(self.step_number)
    def sim(self):
        
        self.v_hist=[]
        for k in range(self.step_number):
            t=self.dt*k


            #アップデートに使用する値を計算
            minf,mt=calculate_m(self.v)
            hinf,ht=calculate_h(self.v)
            ninf,nt=calculate_n(self.v)


            dmdt=(minf-self.m)/mt
            dhdt=(hinf-self.h)/ht
            dndt=(ninf-self.n)/nt



            #アップデート
            self.m=self.m+dmdt*self.dt
            self.h=self.h+dhdt*self.dt
            self.n=self.n+dndt*self.dt

            dvdt= (-self.gna*self.m**3*self.h*(self.v-self.ena) -self.gk*self.n**4*(self.v-self.ek) 
                   -self.gl*(self.v-self.el) + self.stim[k]*self.amp_stim)/self.cm
            self.v=self.v+dvdt*self.dt
            self.v_hist.append(self.v)
    def set_stim(self,flag,interval=0):
        if flag==1:
            self.stim[3300:6600]=1
        elif flag==2:
            sstart=self.step_number/3
            slen=int(1/self.dt)
            self.stim[sstart:sstart+slen]=1
            self.stim[sstart+interval:sstart+interval+slen]=1
        
    def plot(self):
        plt.plot(np.arange(0,self.T,self.dt),self.v_hist)
        plt.ylim([-100,50])
```

```python
#mhnの定常状態（初期状態）での値
h=HHmodel()
print(h.m)
print(h.h)
print(h.n)
```

```python
#33ms~66msの間だけ刺激
h=HHmodel()
h.set_stim(flag=1)
h.sim()
h.plot()
```

```python
#２連続の刺激 不応期
plt.figure(figsize=(15, 6))
interval=[500,1000,2000]

for k in range(3):
    
    h=HHmodel()
    h.set_stim(2,interval[k])
    h.sim()
    plt.subplot(2,3,k+1)
    h.plot()
    plt.title('membrane potential   interval={0}ms'.format(interval[k]*h.dt))
    plt.subplot(2,3,k+4)
    plt.plot(np.arange(0,h.T,h.dt),h.stim)
    plt.ylim([-0.3,1.3])
    plt.title('stimulus timing   interval={0}ms'.format(interval[k]*h.dt))
plt.tight_layout()
```
