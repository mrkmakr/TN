```python
image = Image.open('lenna.jpg')
image=image.convert('L')
img=np.asarray(image)
```

```python
#画像の読み込みと表示
plt.imshow(img)
plt.axis('off')
```

```python
#フィルターの適用
freq=0.4
sigma_x=1.5
sigma_y=1.5
kernel = np.real(gabor_kernel(freq, theta=np.pi,sigma_x=sigma_x, sigma_y=sigma_y))
plt.figure(num=None, figsize=(6, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(2,2,1)
plt.imshow(kernel,interpolation='none')
plt.axis('off')


img_filtered_v=sp.signal.convolve2d(img,kernel,boundary='symm',mode='same')
plt.subplot(2,2,2)
plt.imshow(img_filtered_v)
plt.axis('off')


kernel = np.real(gabor_kernel(freq, theta=np.pi*0.25,sigma_x=sigma_x, sigma_y=sigma_y))
plt.subplot(2,2,3)
plt.imshow(kernel)
plt.axis('off')


img_filtered_h=sp.signal.convolve2d(img,kernel,boundary='symm',mode='same')
plt.subplot(2,2,4)
plt.imshow(img_filtered_h)
plt.axis('off')
```

```python
#エッジ検出結果

img_filtered=np.zeros(img.shape)
for k in range(4):
    kernel = np.real(gabor_kernel(freq, theta=np.pi*0.25*k,sigma_x=sigma_x, sigma_y=sigma_y))
    img_filtered_tmp=sp.signal.convolve2d(img,kernel,boundary='symm',mode='same')
    img_filtered+=img_filtered_tmp

plt.figure(figsize=[10,10])

plt.subplot(1,2,1)
plt.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_filtered)
plt.axis('off')
```

```python
#おまけ
#ガボールを自分で定義
def gauss(x,y,m,sig):
    return 1/np.sqrt(2*np.pi*sig**2)*np.exp(-(x-m)**2/2/sig**2)*1/np.sqrt(2*np.pi*sig**2)*np.exp(-(y-m)**2/2/sig**2)

#filter_size, rotation , sigma of gauss , frequency of cos , phase of cos
def gb(size,theta,sig,omega,k):
    x=np.arange(-size/2.0,size/2.0,0.1)
    y=np.arange(-size/2.0,size/2.0,0.1)

    xx=ml.repmat(x,len(x),1)
    yy=ml.repmat(y,len(y),1).T

    gau=gauss(xx,yy,0,sig)

    xy_rotation=xx*np.cos(theta)+yy*np.sin(theta)

    wave=np.cos(xy_rotation*omega+k)
    
    return(gau*wave)
```

```python
kernel=gb(5,np.pi/4,1,4,np.pi*0.5)
plt.imshow(kernel)
```
